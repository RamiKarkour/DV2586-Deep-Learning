# AI Resume–Job Matching with Deep Learning

This notebook builds and evaluates an NLP-based resume–job matching system. It compares classical baselines with transformer-based models and a hybrid reranking method.

```python
# Core data and file handling
import ast
import json
import re
from collections import Counter
from pathlib import Path

# Numerical computing and data processing
import numpy as np
import pandas as pd

# Visualization
import matplotlib.pyplot as plt

# Classical NLP and ranking baselines
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.model_selection import train_test_split
from rank_bm25 import BM25Okapi

# Deep learning and transformer models
import torch
from torch.utils.data import DataLoader
from sentence_transformers import SentenceTransformer, InputExample, losses
from sentence_transformers.util import batch_to_device
from transformers import get_linear_schedule_with_warmup

# Progress bars
from tqdm.auto import tqdm

# Project paths
BASE_DIR = Path.cwd().parent if Path.cwd().name == "notebooks" else Path.cwd()

RAW_DIR = BASE_DIR / "data" / "raw"
PROCESSED_DIR = BASE_DIR / "data" / "processed"
OUTPUT_DIR = BASE_DIR / "outputs"

resume_csv = RAW_DIR / "resume-dataset" / "resume" / "Resume.csv"

postings_path = RAW_DIR / "linkedin-job-postings" / "postings.csv"
job_skills_path = RAW_DIR / "linkedin-job-postings" / "jobs" / "job_skills.csv"
skills_path = RAW_DIR / "linkedin-job-postings" / "mappings" / "skills.csv"

print("Resume CSV exists:", resume_csv.exists())
print("Postings exists:", postings_path.exists())
print("Job skills exists:", job_skills_path.exists())
print("Skills mapping exists:", skills_path.exists())
```

```python
if resume_csv.exists():
    resumes = pd.read_csv(resume_csv)

jobs = pd.read_csv(postings_path)
job_skills = pd.read_csv(job_skills_path)
skills = pd.read_csv(skills_path)

print("Resumes shape:", resumes.shape)
print("Jobs shape:", jobs.shape)
print("Job skills shape:", job_skills.shape)
print("Skills shape:", skills.shape)
```

```python
print("Resume columns:")
print(resumes.columns.tolist())

print("\nJob posting columns:")
print(jobs.columns.tolist())

print("\nJob skills columns:")
print(job_skills.columns.tolist())

print("\nSkills columns:")
print(skills.columns.tolist())
```

## 3. Text Cleaning Function

This function removes HTML, URLs, unusual symbols, extra whitespace, and lowercases the text. The same cleaning logic is used for resumes and job descriptions.

```python
def clean_text(text):
    if pd.isna(text):
        return ""
    
    text = str(text)
    text = re.sub(r"<.*?>", " ", text)          # remove HTML tags
    text = re.sub(r"http\S+|www\S+", " ", text) # remove URLs
    text = re.sub(r"[^a-zA-Z0-9+#.\s]", " ", text)
    text = re.sub(r"\s+", " ", text).strip()
    return text.lower()
```

## 4. Clean Resume Data

The resume dataset is reduced to the useful fields: resume ID, resume text, and resume category.

```python
resumes_clean = resumes[["ID", "Resume_str", "Category"]].copy()

resumes_clean = resumes_clean.rename(columns={
    "ID": "resume_id",
    "Resume_str": "resume_text",
    "Category": "resume_category"
})

resumes_clean["resume_text_clean"] = resumes_clean["resume_text"].apply(clean_text)
resumes_clean["resume_category"] = resumes_clean["resume_category"].astype(str).str.strip()

resumes_clean = resumes_clean[
    resumes_clean["resume_text_clean"].str.len() > 100
].reset_index(drop=True)

print(resumes_clean.shape)
resumes_clean.head()
```

## 5. Merge Job Skill Labels

LinkedIn job skill abbreviations are merged with readable skill names, then grouped so each job has a single skill list.

```python
job_skills_named = job_skills.merge(skills, on="skill_abr", how="left")

job_skills_grouped = (
    job_skills_named
    .groupby("job_id")["skill_name"]
    .apply(lambda x: ", ".join(sorted(set(x.dropna().astype(str)))))
    .reset_index()
    .rename(columns={"skill_name": "job_skills"})
)

print(job_skills_grouped.shape)
job_skills_grouped.head()
```

## 6. Clean Job Posting Data

Job titles, descriptions, experience level, work type, and skill fields are combined into one cleaned job text field.

```python
job_cols = [
    "job_id",
    "title",
    "description",
    "formatted_experience_level",
    "formatted_work_type",
    "work_type",
    "skills_desc"
]

jobs_clean = jobs[job_cols].copy()

jobs_clean = jobs_clean.merge(job_skills_grouped, on="job_id", how="left")

jobs_clean["title"] = jobs_clean["title"].fillna("")
jobs_clean["description"] = jobs_clean["description"].fillna("")
jobs_clean["skills_desc"] = jobs_clean["skills_desc"].fillna("")
jobs_clean["job_skills"] = jobs_clean["job_skills"].fillna("")

jobs_clean["job_text"] = (
    jobs_clean["title"].astype(str) + " " +
    jobs_clean["description"].astype(str) + " " +
    jobs_clean["skills_desc"].astype(str) + " " +
    jobs_clean["job_skills"].astype(str)
)

jobs_clean["job_text_clean"] = jobs_clean["job_text"].apply(clean_text)
jobs_clean["title_clean"] = jobs_clean["title"].apply(clean_text)

jobs_clean = jobs_clean[
    jobs_clean["job_text_clean"].str.len() > 100
].drop_duplicates(subset=["job_id"]).reset_index(drop=True)

print(jobs_clean.shape)
jobs_clean.head()
```

## 11. Assign Matched Categories to Jobs

Jobs are matched to categories using stricter title/skill matching to reduce false positives from generic description words.

```python
def contains_keyword(text, keyword):
    text = str(text).lower()
    keyword = str(keyword).lower()
    
    # For multi-word keywords, simple escaped phrase match
    if " " in keyword:
        pattern = r"\b" + re.escape(keyword) + r"\b"
    else:
        pattern = r"\b" + re.escape(keyword) + r"\b"
    
    return re.search(pattern, text) is not None


def match_job_categories_strict(title, skills, category_keywords):
    # Use title and explicit skills only to reduce false matches
    text = f"{title} {skills}".lower()
    matched = []
    
    for category, keywords in category_keywords.items():
        for keyword in keywords:
            if contains_keyword(text, keyword):
                matched.append(category)
                break
                
    return matched


jobs_sample["matched_categories"] = jobs_sample.apply(
    lambda row: match_job_categories_strict(
        row["title_clean"],
        row["job_skills"],
        category_keywords
    ),
    axis=1
)

jobs_sample["num_matched_categories"] = jobs_sample["matched_categories"].apply(len)

print(jobs_sample["num_matched_categories"].value_counts().sort_index())
jobs_sample[["job_id", "title", "job_skills", "matched_categories"]].head(20)
```

## 14. Evaluation Metric Function

The same ranking metrics are used for all methods: Top-K Accuracy, Precision@K, and Mean Reciprocal Rank.

```python
# Evaluate a similarity matrix by checking whether relevant jobs appear near the top of each ranking.
def evaluate_rankings(similarity_matrix, resumes_df, jobs_df, top_k_values=(1, 3, 5, 10)):
    results = {}

    resume_categories = resumes_df["resume_category"].tolist()
    job_categories = jobs_df["matched_categories"].tolist()

    top1_correct = 0
    reciprocal_ranks = []
    precision_at_k = {k: [] for k in top_k_values}
    topk_accuracy = {k: 0 for k in top_k_values}

    for i, resume_category in enumerate(resume_categories):
        scores = similarity_matrix[i]
        ranked_job_indices = np.argsort(scores)[::-1]

        relevance = []
        for job_idx in ranked_job_indices:
            is_relevant = resume_category in job_categories[job_idx]
            relevance.append(is_relevant)

        # Top-K accuracy
        for k in top_k_values:
            if any(relevance[:k]):
                topk_accuracy[k] += 1

            precision_at_k[k].append(sum(relevance[:k]) / k)

        # MRR
        first_relevant_rank = None
        for rank, is_relevant in enumerate(relevance, start=1):
            if is_relevant:
                first_relevant_rank = rank
                break

        if first_relevant_rank is not None:
            reciprocal_ranks.append(1 / first_relevant_rank)
        else:
            reciprocal_ranks.append(0)

    n = len(resumes_df)

    for k in top_k_values:
        results[f"Top-{k} Accuracy"] = topk_accuracy[k] / n
        results[f"Precision@{k}"] = np.mean(precision_at_k[k])

    results["MRR"] = np.mean(reciprocal_ranks)

    return results
```

## 15. Method 1: TF-IDF + Cosine Similarity

This classical baseline represents resumes and job descriptions using TF-IDF vectors, then ranks jobs using cosine similarity.

```python
tfidf_vectorizer = TfidfVectorizer(
    max_features=30000,
    stop_words="english",
    ngram_range=(1, 2),
    min_df=2
)

all_texts = pd.concat([
    resumes_eval["resume_text_clean"],
    jobs_eval["job_text_clean"]
], ignore_index=True)

tfidf_matrix = tfidf_vectorizer.fit_transform(all_texts)

resume_tfidf = tfidf_matrix[:len(resumes_eval)]
job_tfidf = tfidf_matrix[len(resumes_eval):]

tfidf_similarity = cosine_similarity(resume_tfidf, job_tfidf)

print("TF-IDF similarity matrix shape:", tfidf_similarity.shape)
```

```python
tfidf_results = evaluate_rankings(
    tfidf_similarity,
    resumes_eval,
    jobs_eval,
    top_k_values=(1, 3, 5, 10)
)

tfidf_results
```

```python
METRICS_DIR = OUTPUT_DIR / "metrics"
METRICS_DIR.mkdir(parents=True, exist_ok=True)

tfidf_metrics_df = pd.DataFrame([{
    "method": "TF-IDF + Cosine Similarity",
    **tfidf_results
}])

tfidf_metrics_df.to_csv(METRICS_DIR / "tfidf_metrics.csv", index=False)

tfidf_metrics_df
```

```python
RESULTS_DIR = OUTPUT_DIR / "results"
RESULTS_DIR.mkdir(parents=True, exist_ok=True)

def get_top_matches(similarity_matrix, resume_index, resumes_df, jobs_df, top_n=5):
    scores = similarity_matrix[resume_index]
    top_indices = np.argsort(scores)[::-1][:top_n]

    rows = []
    resume_category = resumes_df.iloc[resume_index]["resume_category"]
    
    for rank, job_idx in enumerate(top_indices, start=1):
        job = jobs_df.iloc[job_idx]
        rows.append({
            "rank": rank,
            "resume_id": resumes_df.iloc[resume_index]["resume_id"],
            "resume_category": resume_category,
            "job_id": job["job_id"],
            "job_title": job["title"],
            "job_categories": ", ".join(job["matched_categories"]),
            "score": scores[job_idx],
            "relevant": resume_category in job["matched_categories"]
        })
    
    return pd.DataFrame(rows)

sample_rankings = pd.concat([
    get_top_matches(tfidf_similarity, i, resumes_eval, jobs_eval, top_n=5)
    for i in range(min(10, len(resumes_eval)))
], ignore_index=True)

sample_rankings.to_csv(RESULTS_DIR / "tfidf_sample_rankings.csv", index=False)

sample_rankings
```

## 16. Method 2: BM25 Ranking

BM25 is a search-engine-style information retrieval baseline. It ranks job postings according to their term relevance to each resume query.

# BM25

```python
def tokenize_text(text):
    return str(text).lower().split()

job_corpus_tokens = jobs_eval["job_text_clean"].apply(tokenize_text).tolist()
resume_query_tokens = resumes_eval["resume_text_clean"].apply(tokenize_text).tolist()

bm25 = BM25Okapi(job_corpus_tokens)

print("BM25 corpus size:", len(job_corpus_tokens))
print("Number of resume queries:", len(resume_query_tokens))
```

```python
bm25_scores = []

for query_tokens in tqdm(resume_query_tokens, desc="Computing BM25 scores"):
    scores = bm25.get_scores(query_tokens)
    bm25_scores.append(scores)

bm25_similarity = np.vstack(bm25_scores)

print("BM25 similarity matrix shape:", bm25_similarity.shape)
```

```python
bm25_results = evaluate_rankings(
    bm25_similarity,
    resumes_eval,
    jobs_eval,
    top_k_values=(1, 3, 5, 10)
)

bm25_results
```

```python
bm25_metrics_df = pd.DataFrame([{
    "method": "BM25",
    **bm25_results
}])

bm25_metrics_df.to_csv(METRICS_DIR / "bm25_metrics.csv", index=False)

bm25_metrics_df
```

```python
bm25_sample_rankings = pd.concat([
    get_top_matches(bm25_similarity, i, resumes_eval, jobs_eval, top_n=5)
    for i in range(min(10, len(resumes_eval)))
], ignore_index=True)

bm25_sample_rankings.to_csv(RESULTS_DIR / "bm25_sample_rankings.csv", index=False)

bm25_sample_rankings
```

```python
baseline_comparison = pd.concat(
    [tfidf_metrics_df, bm25_metrics_df],
    ignore_index=True
)

baseline_comparison.to_csv(METRICS_DIR / "baseline_comparison.csv", index=False)

baseline_comparison
```

## 18. Method 3: Pretrained Sentence-BERT

A pretrained Sentence-BERT model converts resumes and job postings into dense semantic embeddings. Jobs are ranked using cosine similarity between embeddings.

# Pretrained Sentence-BERT + cosine similarity

```python
device = "cuda" if torch.cuda.is_available() else "cpu"
print("Using device:", device)

model = SentenceTransformer("all-MiniLM-L6-v2", device=device)
```

```python
resume_embeddings = model.encode(
    resumes_eval["resume_text_clean"].tolist(),
    batch_size=64,
    convert_to_numpy=True,
    show_progress_bar=True,
    device=device
)

job_embeddings = model.encode(
    jobs_eval["job_text_clean"].tolist(),
    batch_size=64,
    convert_to_numpy=True,
    show_progress_bar=True,
    device=device
)

print("Resume embeddings shape:", resume_embeddings.shape)
print("Job embeddings shape:", job_embeddings.shape)
```

```python
sbert_similarity = cosine_similarity(resume_embeddings, job_embeddings)

print("SBERT similarity matrix shape:", sbert_similarity.shape)
```

```python
sbert_results = evaluate_rankings(
    sbert_similarity,
    resumes_eval,
    jobs_eval,
    top_k_values=(1, 3, 5, 10)
)

sbert_results
```

## 20. Create Positive Training Pairs

Positive resume–job pairs are created when the resume category appears in a job's matched categories.

```python
rng = np.random.default_rng(RANDOM_STATE)

category_to_job_indices = {}

for category in valid_categories:
    indices = jobs_eval[
        jobs_eval["matched_categories"].apply(lambda cats: category in cats)
    ].index.tolist()
    
    category_to_job_indices[category] = indices

print("Jobs available per category:")
for category, indices in category_to_job_indices.items():
    print(category, len(indices))
```

```python
# For each training resume, sample multiple relevant job postings from the same category.
PAIRS_PER_RESUME = 3

train_examples = []

for _, resume_row in train_resumes.iterrows():
    category = resume_row["resume_category"]
    resume_text = resume_row["resume_text_clean"]
    
    possible_job_indices = category_to_job_indices.get(category, [])
    
    if len(possible_job_indices) == 0:
        continue
    
    sampled_job_indices = rng.choice(
        possible_job_indices,
        size=min(PAIRS_PER_RESUME, len(possible_job_indices)),
        replace=len(possible_job_indices) < PAIRS_PER_RESUME
    )
    
    for job_idx in sampled_job_indices:
        job_text = jobs_eval.loc[job_idx, "job_text_clean"]
        train_examples.append(InputExample(texts=[resume_text, job_text]))

print("Number of training pairs:", len(train_examples))
print("Example pair:")
print(train_examples[0].texts[0][:300])
print("-----")
print(train_examples[0].texts[1][:300])
```

## 21. Method 4: Fine-tuned Sentence-BERT

This is the main deep learning method. Sentence-BERT is fine-tuned using Multiple Negatives Ranking Loss so matching resume–job pairs move closer in embedding space.

```python
device = "cuda" if torch.cuda.is_available() else "cpu"
print("Using device:", device)

model_ft = SentenceTransformer("all-MiniLM-L6-v2", device=device)

TRAIN_BATCH_SIZE = 16

train_dataloader = DataLoader(
    train_examples,
    shuffle=True,
    batch_size=TRAIN_BATCH_SIZE,
    collate_fn=model_ft.smart_batching_collate
)

train_loss = losses.MultipleNegativesRankingLoss(model_ft)

print("Training batches per epoch:", len(train_dataloader))
```

```python
# Fine-tuning hyperparameters.
EPOCHS = 2
LEARNING_RATE = 2e-5
WARMUP_RATIO = 0.1

optimizer = torch.optim.AdamW(model_ft.parameters(), lr=LEARNING_RATE)

total_training_steps = len(train_dataloader) * EPOCHS
warmup_steps = int(total_training_steps * WARMUP_RATIO)

scheduler = get_linear_schedule_with_warmup(
    optimizer,
    num_warmup_steps=warmup_steps,
    num_training_steps=total_training_steps
)

loss_history = []

model_ft.train()

for epoch in range(EPOCHS):
    progress_bar = tqdm(train_dataloader, desc=f"Epoch {epoch + 1}/{EPOCHS}")
    
    for step, batch in enumerate(progress_bar):
        features, labels = batch
        
        features = [batch_to_device(feature, device) for feature in features]
        labels = labels.to(device)
        
        optimizer.zero_grad()
        
        loss_value = train_loss(features, labels)
        loss_value.backward()
        
        torch.nn.utils.clip_grad_norm_(model_ft.parameters(), max_norm=1.0)
        
        optimizer.step()
        scheduler.step()
        
        current_loss = loss_value.item()
        
        loss_history.append({
            "epoch": epoch + 1,
            "step": step + 1,
            "loss": current_loss
        })
        
        progress_bar.set_postfix({"loss": current_loss})

loss_history_df = pd.DataFrame(loss_history)
loss_history_df.to_csv(METRICS_DIR / "finetuned_sbert_training_loss.csv", index=False)

loss_history_df.tail()
```

## 25. Method 5: Hybrid Reranking

The hybrid method combines fine-tuned SBERT similarity with explicit skill-overlap scores.

# Hybrid Reranking

```python
def extract_skills_from_text(text, skill_names):
    text = str(text).lower()
    found_skills = set()
    
    for skill in skill_names:
        skill_lower = str(skill).lower()
        pattern = r"\b" + re.escape(skill_lower) + r"\b"
        if re.search(pattern, text):
            found_skills.add(skill_lower)
            
    return found_skills


# Use the skill names from the LinkedIn skill mapping file
skill_names = skills["skill_name"].dropna().astype(str).unique().tolist()

print("Number of skill names:", len(skill_names))
print(skill_names[:20])
```

```python
test_resume_skill_sets = [
    extract_skills_from_text(text, skill_names)
    for text in test_resumes["resume_text_clean"].tolist()
]

job_skill_sets = [
    set(str(skills_text).lower().split(", ")) if pd.notna(skills_text) and str(skills_text).strip() != "" else set()
    for skills_text in jobs_eval["job_skills"].tolist()
]

print("Example resume skills:", list(test_resume_skill_sets[0])[:10])
print("Example job skills:", list(job_skill_sets[0])[:10])
```

```python
skill_overlap_matrix = np.zeros((len(test_resumes), len(jobs_eval)))

for i, resume_skills in enumerate(test_resume_skill_sets):
    for j, job_skills_set in enumerate(job_skill_sets):
        if len(resume_skills) == 0 or len(job_skills_set) == 0:
            skill_overlap_matrix[i, j] = 0.0
        else:
            intersection = len(resume_skills.intersection(job_skills_set))
            union = len(resume_skills.union(job_skills_set))
            skill_overlap_matrix[i, j] = intersection / union

print("Skill overlap matrix:", skill_overlap_matrix.shape)
print("Min:", skill_overlap_matrix.min())
print("Max:", skill_overlap_matrix.max())
print("Mean:", skill_overlap_matrix.mean())
```

```python
def min_max_normalize(matrix):
    min_val = matrix.min()
    max_val = matrix.max()
    
    if max_val - min_val == 0:
        return np.zeros_like(matrix)
    
    return (matrix - min_val) / (max_val - min_val)

finetuned_similarity_norm = min_max_normalize(finetuned_similarity)
skill_overlap_norm = min_max_normalize(skill_overlap_matrix)

print("Fine-tuned norm:", finetuned_similarity_norm.min(), finetuned_similarity_norm.max())
print("Skill overlap norm:", skill_overlap_norm.min(), skill_overlap_norm.max())
```

```python
# Final hybrid setting selected from the ablation study.
ALPHA = 0.85
BETA = 0.15

hybrid_similarity = (ALPHA * finetuned_similarity_norm) + (BETA * skill_overlap_norm)

hybrid_results = evaluate_rankings(
    hybrid_similarity,
    test_resumes,
    jobs_eval,
    top_k_values=(1, 3, 5, 10)
)

hybrid_metrics_df = pd.DataFrame([{
    "method": "Hybrid Fine-tuned SBERT + Skill Overlap",
    **hybrid_results
}])

hybrid_metrics_df
```

```python
final_comparison_with_hybrid = pd.concat(
    [
        tfidf_test_metrics_df,
        bm25_test_metrics_df,
        pretrained_test_metrics_df,
        finetuned_metrics_df,
        hybrid_metrics_df
    ],
    ignore_index=True
)

final_comparison_with_hybrid.to_csv(
    METRICS_DIR / "final_model_comparison_with_hybrid.csv",
    index=False
)

final_comparison_with_hybrid
```

## 26. Hybrid Weight Ablation

Different SBERT/skill weights are tested to find the best hybrid configuration.

# alpha/beta ablation

```python
alpha_values = [0.70, 0.75, 0.80, 0.85, 0.90, 0.95]

hybrid_ablation_results = []

for alpha in alpha_values:
    beta = 1 - alpha
    
    hybrid_scores = (alpha * finetuned_similarity_norm) + (beta * skill_overlap_norm)
    
    results = evaluate_rankings(
        hybrid_scores,
        test_resumes,
        jobs_eval,
        top_k_values=(1, 3, 5, 10)
    )
    
    hybrid_ablation_results.append({
        "alpha_sbert": alpha,
        "beta_skill_overlap": beta,
        **results
    })

hybrid_ablation_df = pd.DataFrame(hybrid_ablation_results)

hybrid_ablation_df.to_csv(
    METRICS_DIR / "hybrid_weight_ablation.csv",
    index=False
)

hybrid_ablation_df
```

## 28. Qualitative and Failure Analysis

Top-ranked examples are saved to analyze correct predictions, failure cases, and cases where the hybrid method improved or worsened results.

# Helper function for detailed top-k rankings

```python
def get_detailed_top_matches(similarity_matrix, resume_index, resumes_df, jobs_df, method_name, top_n=5):
    scores = similarity_matrix[resume_index]
    top_indices = np.argsort(scores)[::-1][:top_n]

    resume_row = resumes_df.iloc[resume_index]
    resume_category = resume_row["resume_category"]

    rows = []

    for rank, job_idx in enumerate(top_indices, start=1):
        job = jobs_df.iloc[job_idx]

        rows.append({
            "method": method_name,
            "resume_index": resume_index,
            "rank": rank,
            "resume_id": resume_row["resume_id"],
            "resume_category": resume_category,
            "resume_text_preview": resume_row["resume_text_clean"][:300],
            "job_id": job["job_id"],
            "job_title": job["title"],
            "job_categories": ", ".join(job["matched_categories"]),
            "job_skills": job["job_skills"],
            "score": scores[job_idx],
            "relevant": resume_category in job["matched_categories"]
        })

    return pd.DataFrame(rows)
```

```python
final_ranking_examples = pd.concat([
    get_detailed_top_matches(finetuned_similarity, i, test_resumes, jobs_eval, "Fine-tuned Sentence-BERT", top_n=5)
    for i in range(len(test_resumes))
], ignore_index=True)

hybrid_ranking_examples = pd.concat([
    get_detailed_top_matches(hybrid_similarity, i, test_resumes, jobs_eval, "Hybrid Fine-tuned SBERT + Skill Overlap", top_n=5)
    for i in range(len(test_resumes))
], ignore_index=True)

final_ranking_examples.to_csv(
    RESULTS_DIR / "finetuned_sbert_all_test_rankings_top5.csv",
    index=False
)

hybrid_ranking_examples.to_csv(
    RESULTS_DIR / "hybrid_all_test_rankings_top5.csv",
    index=False
)

print("Fine-tuned rankings:", final_ranking_examples.shape)
print("Hybrid rankings:", hybrid_ranking_examples.shape)
```

```python
finetuned_top1 = final_ranking_examples[
    final_ranking_examples["rank"] == 1
][["resume_index", "resume_category", "job_title", "job_categories", "relevant"]].copy()

finetuned_top1 = finetuned_top1.rename(columns={
    "job_title": "finetuned_top1_job",
    "job_categories": "finetuned_top1_categories",
    "relevant": "finetuned_top1_relevant"
})

hybrid_top1_compare = hybrid_top1[
    ["resume_index", "resume_category", "job_title", "job_categories", "relevant"]
].copy()

hybrid_top1_compare = hybrid_top1_compare.rename(columns={
    "job_title": "hybrid_top1_job",
    "job_categories": "hybrid_top1_categories",
    "relevant": "hybrid_top1_relevant"
})

top1_comparison = finetuned_top1.merge(
    hybrid_top1_compare,
    on=["resume_index", "resume_category"],
    how="inner"
)

hybrid_improved_cases = top1_comparison[
    (top1_comparison["finetuned_top1_relevant"] == False) &
    (top1_comparison["hybrid_top1_relevant"] == True)
].copy()

hybrid_worse_cases = top1_comparison[
    (top1_comparison["finetuned_top1_relevant"] == True) &
    (top1_comparison["hybrid_top1_relevant"] == False)
].copy()

hybrid_improved_cases.to_csv(
    RESULTS_DIR / "hybrid_improved_over_finetuned_cases.csv",
    index=False
)

hybrid_worse_cases.to_csv(
    RESULTS_DIR / "hybrid_worse_than_finetuned_cases.csv",
    index=False
)

print("Hybrid improved cases:", len(hybrid_improved_cases))
print("Hybrid worse cases:", len(hybrid_worse_cases))

hybrid_improved_cases.head(10)
```

```python
def top1_correct_by_category(similarity_matrix, resumes_df, jobs_df):
    rows = []

    for i, resume_row in resumes_df.iterrows():
        scores = similarity_matrix[i]
        top_idx = np.argmax(scores)

        resume_category = resume_row["resume_category"]
        job_categories = jobs_df.iloc[top_idx]["matched_categories"]

        rows.append({
            "resume_category": resume_category,
            "top1_correct": resume_category in job_categories
        })

    result_df = pd.DataFrame(rows)

    category_perf = (
        result_df
        .groupby("resume_category")["top1_correct"]
        .agg(["count", "mean"])
        .reset_index()
        .rename(columns={
            "count": "num_test_resumes",
            "mean": "top1_accuracy"
        })
        .sort_values("top1_accuracy", ascending=False)
    )

    return category_perf


hybrid_category_performance = top1_correct_by_category(
    hybrid_similarity,
    test_resumes,
    jobs_eval
)

hybrid_category_performance.to_csv(
    METRICS_DIR / "hybrid_per_category_top1_accuracy.csv",
    index=False
)

hybrid_category_performance
```
